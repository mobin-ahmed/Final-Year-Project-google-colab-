### equation of circle:
1. x2+y2=r2 where center is (0,0)
2. (x-h)2+(y-k)2=r2 where center is (h,k)
3. draw the circle by using mid point circle algorithm

### input the radius of the circle and the origin/center of the circle is=(h,k)

```python
r=100
h0,k0=100,100
```

### full mid point circle algorithm code

```python
h,k=0,0
arr=[]
arr2=[]
xk,yk=h,r
arr.append([xk,yk])
p=1-r

while True:
    if xk==yk:
        arr2=arr.copy()
        arr.pop()
        for i in range(len(arr)):
            x=arr.pop()
            arr2.append([x[1],x[0]])
        break
    if xk>yk:
        arr2=arr.copy()
        arr2.pop()
        arr.pop()
        for i in range(len(arr)):
            x=arr.pop()
            arr2.append([x[1],x[0]])
        break   
    if p<0:
        p=p+2*xk+3
        xk=xk+1
        yk=yk
        arr.append([xk,yk])
    elif p>=0:
        p=p+2*(xk-yk)+5
        xk=xk+1
        yk=yk-1
        arr.append([xk,yk])

        
arr=arr2.copy()
arr2.pop()
l=int(len(arr2))
for i in range(l):
    x=arr2.pop()
    arr.append([x[0],-x[1]])

    
x=arr.pop()
arr2=arr.copy()
arr2.append(x)
while len(arr):
    x=arr.pop()
    arr2.append([-x[0],x[1]])
    if x[0]==r and x[1]==h:
        break

        
while len(arr):
    if x[0]==h and x[1]==r:
        arr2.append([x[0],x[1]])
        break
    x=arr.pop()
    arr2.append([-x[0],x[1]])
    
    
if h0 is not 0 and k0 is not 0:
    for i in arr2:
        i[0]+=h0
        i[1]+=k0
```
