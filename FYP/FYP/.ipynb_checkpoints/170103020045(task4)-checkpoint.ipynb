### plotting the mean shape of the Alphabet "a":

```python
print(len(mean_arr_x))
print(len(mean_arr_y))
```

```python
marrx=[]
marry=[]
v=0
while True:
    sm1,sm2=0,0
    for i in range(len(mean_arr_x)):
        sm1+=mean_arr_x[i][v]
        sm2+=mean_arr_y[i][v]
#     print(sm1,sm2)
    marrx.append(sm1/len(mean_arr_x))
    marry.append(sm2/len(mean_arr_x))
    v+=1
    if v==len(mean_arr_x[i]):
        break
```

```python
print(marrx)
print(marry)
```

```python
print(len(marrx))
print(len(marry))
print(v)
```

```python
for i in range(len(marrx)):
    marrx[i]=(marrx[i]*150)+100
    marry[i]=(marry[i]*60)+50
```

```python
print(marrx)
print(marry)
```

```python
plt.scatter(marrx,marry)
plt.plot(marrx,marry)
```

```python
bot=AutoKiddobot()
```

```python
bot.start()
```

```python
bot.go_to_xy(marrx[0],marry[0])
```

```python
bot.pen_down()
```

```python
for i,j in zip(marrx,marry):
    bot.go_to_xy(i,j)
```

```python
bot.pen_up()
```

```python
bot.close()
```
